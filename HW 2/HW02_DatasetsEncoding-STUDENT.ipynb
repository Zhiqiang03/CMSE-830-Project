```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
from cycler import cycler

# Generate sample data
x = np.linspace(0, 10, 100) # 100 points from 0 to 10
y1 = np.sin(x) # sine values at those points
y2 = np.cos(x) # cosine values at those points
y3 = np.tan(x) # tangent values at those points

# Create a 3x3 grid of subplots with different styles
fig, axs = plt.subplots(3, 3, figsize=(18, 18)) # a big figure with 9 subplots (3 rows, 3 columns)
fig.suptitle('Matplotlib Style and Customization Demo', fontsize=20) # overall title for the figure

# Function to set common properties for all subplots
def set_common_properties(ax, title):
    ax.set_title(title) # set the title of the subplot
    ax.set_xlabel('x') # set the x-axis label
    ax.set_ylabel('y') # set the y-axis label
    ax.legend() # show the legend

# Plot 1: Default style
axs[0, 0].plot(x, y1, label='sin(x)') # plot sine function at subplot (0,0)
axs[0, 0].plot(x, y2, label='cos(x)') # plot cosine function at subplot (0,0)
set_common_properties(axs[0, 0], 'Default Style') # set common properties for this subplot
axs[0, 0].grid(True, linestyle='--', alpha=0.7) # add grid with dashed lines and some transparency

# Plot 2: Seaborn style
with plt.style.context('fivethirtyeight'): # use fivethirtyeight style for this block
    axs[0, 1].plot(x, y1, label='sin(x)') # plot sine function at subplot (0,1)
    axs[0, 1].plot(x, y2, label='cos(x)') # plot cosine function at subplot (0,1)
    set_common_properties(axs[0, 1], '538') # set common properties for this subplot

# Plot 3: ggplot style
with plt.style.context('ggplot'): # use ggplot style for this block
    axs[0, 2].plot(x, y1, label='sin(x)') # plot sine function at subplot (0,2)
    axs[0, 2].plot(x, y2, label='cos(x)') # plot cosine function at subplot (0,2)
    set_common_properties(axs[0, 2], 'ggplot Style') # set common properties for this subplot

# Plot 4: Custom colors and markers
axs[1, 0].plot(x, y1, color='#FF5733', marker='o', markersize=4, label='sin(x)') # plot sine with custom color and circle markers
axs[1, 0].plot(x, y2, color='#33FF57', marker='s', markersize=4, label='cos(x)') # plot cosine with custom color and square markers
set_common_properties(axs[1, 0], 'Custom Colors and Markers') #set common properties for this subplot
axs[1, 0].set_facecolor('#F0F0F0') # set background color for this subplot

# Plot 5: Filled areas
axs[1, 1].fill_between(x, y1, alpha=0.3, label='sin(x) area') # fill area under sine curve with transparency
axs[1, 1].fill_between(x, y2, alpha=0.3, label='cos(x) area') # fill area under cosine curve with transparency
set_common_properties(axs[1, 1], 'Filled Areas') #set common properties for this subplot

# Plot 6: Colorblind-friendly palette
with plt.style.context('seaborn-v0_8-colorblind'): # use seaborn colorblind style for this block
    # Use the 'colorblind' color cycle from seaborn
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=plt.cm.Set2.colors) # set color cycle to Set2 colormap
    
    for y, label in zip([y1, y2, y3], ['sin(x)', 'cos(x)', 'tan(x)']): # plot sine, cosine, and tangent functions
        axs[1, 2].plot(x, y, label=label) # plot each function with its label
    set_common_properties(axs[1, 2], 'Colorblind-friendly (Seaborn)') #set common properties for this subplot

# Plot 7: Log scale
axs[2, 0].semilogy(x, np.abs(y1), label='|sin(x)|') # plot abs sine function on a semilogarithmic scale
axs[2, 0].semilogy(x, np.abs(y2), label='|cos(x)|') # plot abs cosine function on a semilogarithmic scale
set_common_properties(axs[2, 0], 'Log Scale') #set common properties for this subplot

# Plot 8: Horizontal bar chart
languages = ['Python', 'Java', 'C++', 'JavaScript', 'R'] # list of programming languages
popularity = [68, 45, 35, 62, 25] # corresponding popularity scores
axs[2, 1].barh(languages, popularity) # horizontal bar chart of language popularity
axs[2, 1].set_title('Horizontal Bar Chart') #set title for this subplot
axs[2, 1].set_xlabel('Popularity') #set x-axis label
axs[2, 1].set_ylabel('Programming Language') #set y-axis label

# Plot 9: Scatter plot with size and color mapping
n = 50 # number of points
x = np.random.rand(n) # random x values
y = np.random.rand(n) # random y values
colors = np.random.rand(n) # random colors
sizes = 1000 * np.random.rand(n) # random sizes scaled up

scatter = axs[2, 2].scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis') # scatter plot with color and size mapping
axs[2, 2].set_title('Scatter with Size and Color Mapping') #set title for this subplot
axs[2, 2].set_xlabel('X values') #set x-axis label
axs[2, 2].set_ylabel('Y values') #set y-axis label
plt.colorbar(scatter, ax=axs[2, 2], label='Color Value') # add color bar to indicate color values

# Adjust layout and display the plot
plt.tight_layout() # adjust layout to prevent overlap
plt.show() # display the figure with all subplots
```

## Part 4. Scaling Data (10 points)

üóíÔ∏è **Task:** In this problem you will examine a very popular dataset and scale it. Your only job: comment every line of this code and explain what it does. 

```python
import numpy as np  # Import NumPy for numerical array operations.
import matplotlib.pyplot as plt  # Import Matplotlib for plotting.
from sklearn.datasets import load_iris  # Import utility to load the Iris dataset.
from sklearn.preprocessing import StandardScaler  # Import feature scaler for standardization.

plt.rcdefaults() # reset Matplotlib config

iris = load_iris() # Load the iris dataset
X = iris.data # get the data
feature_names = iris.feature_names # get the feature names

scaler = StandardScaler() # create a scaler instance

X_scaled = scaler.fit_transform(X) # fit x use scaler

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))  #Create first figure with two side-by-side subplots.


ax1.scatter(X[:, 0], X[:, 1], alpha=0.5)  # Scatter plot of first two original features.
ax1.set_title('Original Data')  # Title for original data subplot.
ax1.set_xlabel(feature_names[0])  # X-axis label using original feature name.
ax1.set_ylabel(feature_names[1])  # Y-axis label using original feature name.

ax2.scatter(X_scaled[:, 0], X_scaled[:, 1], alpha=0.5)  # Scatter plot of first two scaled features.
ax2.set_title('Scaled Data')  # Title for scaled data subplot.
ax2.set_xlabel(f'Scaled {feature_names[0]}')  # X-axis label indicating scaling applied.
ax2.set_ylabel(f'Scaled {feature_names[1]}')  # Y-axis label indicating scaling applied.

plt.tight_layout()  # Adjust layout to prevent label overlap.

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))  # Create second figure for remaining two features.

ax1.scatter(X[:, 2], X[:, 3], alpha=0.5)  # Scatter plot of third vs fourth original features.
ax1.set_title('Original Data')  # Title for original data subplot.
ax1.set_xlabel(feature_names[2])  # X-axis label for third feature.
ax1.set_ylabel(feature_names[3])  # Y-axis label for fourth feature.

ax2.scatter(X_scaled[:, 2], X_scaled[:, 3], alpha=0.5)  # Scatter plot of third vs fourth scaled features.
ax2.set_title('Scaled Data')  # Title for scaled data subplot.
ax2.set_xlabel(f'Scaled {feature_names[2]}')  # X-axis label indicating scaling.
ax2.set_ylabel(f'Scaled {feature_names[3]}')  # Y-axis label indicating scaling.

plt.tight_layout()  # Adjust layout again for the second figure.

print("Original data:")  # print
print("Means:", np.mean(X, axis=0))  # print the mean
print("Standard deviations:", np.std(X, axis=0))  # print the deviation

print("\nScaled data:")  # print
print("Means:", np.mean(X_scaled, axis=0))  # print scaled mean
print("Standard deviations:", np.std(X_scaled, axis=0))  # print scaled deviation

X_reconstructed = scaler.inverse_transform(X_scaled)  # Invert scaling to reconstruct original feature values.

max_diff = np.max(np.abs(X - X_reconstructed))  # Compute maximum absolute reconstruction error.
print(f"\nMaximum difference between original and reconstructed data: {max_diff}")  # Report reconstruction fidelity.

# Homework Task: Comment on every line of this code, explaining what it does and why it's important.
```

```python
### ANSWER

oe = OrdinalEncoder()
df['Pclass_ordinal'] = oe.fit_transform(df[['Pclass']])
df[['Pclass', 'Pclass_ordinal']]

# Ordinal encoding converts categorical data into integers, specifically for variables that have a natural, ranked order. In this output, the unique values in the Pclass column (1, 2, 3) is map to the integers 0.0, 1.0, and 2.0. this method is very useful because it will preserve the rank of the data.
```

```python
### ANSWER

def custom_binary_encode(value):
    return format(value, '02b')

df['Pclass_binary'] = df['Pclass'].apply(custom_binary_encode)
print(df[['Pclass', 'Pclass_binary']].head())

# this is a binary encode it will change the data value to its binary format, in the output, the passenger class 1 is converted to '01', 2 becomes '10', and 3 becomes '11'. This method is memory efficient, since it create fewer features compare to one hot encoding.
```

```python
### ANSWER

freq_encoding = df.groupby('Pclass').size()/len(df)
df['Pclass_freq'] = df['Pclass'].map(freq_encoding)
print(df[['Pclass', 'Pclass_freq']].head())

# this is frequency encoding, it will replace the data values with their frequency value. it will count the people in each category and the devided by total people. this method is useful for getting the relative frequency, but if all categories have the same frequency it will be hard to distinguish them.
```
