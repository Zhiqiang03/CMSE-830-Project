```python
import pandas as pd
import os

#list files in the weather data directory
weather_path = os.listdir("./data/weather")
```

```python
#append all csv files in the weather data directory into a single dataframe

weather_df = pd.DataFrame()
for file in weather_path:
    if file.endswith(".csv"):
        temp_df = pd.read_csv(f"./data/weather/{file}")
        weather_df = pd.concat([weather_df, temp_df], ignore_index=True)
```

```python
weather_df
```

```python
#delete name column
weather_df = weather_df.drop(columns=['name'])
```

```python
weather_df['datetime'] = pd.to_datetime(weather_df['datetime'])
```

```python
weather_df
```

```python
#cheange preciptype column to rain or not boolean
weather_df['raining'] = weather_df['preciptype'].notnull()
weather_df = weather_df.drop(columns=['preciptype'])
```

```python
#save to csv
weather_df.to_csv('./data/weather_data.csv', index=False)
```
