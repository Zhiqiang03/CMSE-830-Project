```python
import matplotlib.pyplot as plt
import numpy as np

# Customize rcParams
plt.rcParams.update({
    # Remove top and right spines
    'axes.spines.top': False,
    'axes.spines.right': False,
    
    # Increase font size for axis labels
    'axes.labelsize': 14,
    'xtick.labelsize': 12,
    'ytick.labelsize': 12,
    
    # Add a light gray grid with low alpha
    'axes.grid': True,
    'grid.color': 'gray',
    'grid.alpha': 0.1,
    
    # Increase the default figure size
    'figure.figsize': [10, 6],
    
    # Use a higher DPI for better quality
    'figure.dpi': 120,
    
    # Set a global font family
    'font.family': 'sans-serif',
    'font.sans-serif': ['Arial', 'DejaVu Sans', 'Liberation Sans', 'Bitstream Vera Sans', 'sans-serif'],
    
    # Increase the linewidth of the axes
    'axes.linewidth': 1.5,
    
    # Customize the tick marks
    'xtick.major.size': 5,
    'ytick.major.size': 5,
    'xtick.major.width': 1.5,
    'ytick.major.width': 1.5,
    
    # Use LaTeX for math rendering (if available)
    'text.usetex': False,  # Set to True if LaTeX is installed
    
    # Adjust padding to prevent clipping of tick-labels
    'axes.autolimit_mode': 'round_numbers',
    'axes.xmargin': 0.1,
    'axes.ymargin': 0.1,
    
    # Use a color-blind friendly palette
    'axes.prop_cycle': plt.cycler('color', ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB'])
})

# Function to add grid with specified alpha
def add_grid(ax, alpha=0.1):
    # pass
    ax.grid(True, color='gray', alpha=alpha, linestyle='--')

# Example plots
x = np.linspace(0, 10, 1000)

fig, ax = plt.subplots()
# Plot sine waves with different frequencies
frequencies = [1, 2, 3, 4]
for freq in frequencies:
    y = np.sin(freq * x)
    ax.plot(x, y, label=f'sin({freq}x)')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Sine Waves at Different Frequencies', fontsize=16)
ax.legend()
add_grid(ax, alpha=0.3)  # Add grid with alpha=0.3

plt.tight_layout()
plt.show()

# Example of a more complex plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 4))

# Upper plot: sine and cosine
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.plot(x, np.cos(x), label='cos(x)')
ax1.set_xlabel('X-axis')
ax1.set_ylabel('Y-axis')
ax1.set_title('Sine and Cosine', fontsize=16)
ax1.legend()
add_grid(ax1, alpha=0.05)  # Add grid with alpha=0.05

# Lower plot: exponential decay
ax2.plot(x, np.exp(-x), label='exp(-x)')
ax2.plot(x, np.exp(-0.5*x), label='exp(-0.5x)')
ax2.plot(x, np.exp(-0.2*x), label='exp(-0.2x)')
ax2.set_xlabel('X-axis')
ax2.set_ylabel('Y-axis')
ax2.set_title('Exponential Decay', fontsize=16)
ax2.legend()
add_grid(ax2, alpha=0.1)  # Add grid with alpha=0.1

plt.tight_layout()
plt.show()
```

<div style="border: 1px solid #4CAF50; padding: 10px; border-radius: 5px;">
    <h2 style="color: #4CAF50;">Key Polyphenols in Wine</h2>
    <p>Let's explore the most important polyphenols in wine that contribute to its health benefits and flavor profile:</p>
    <ul>
        <li><strong>Resveratrol</strong>: Found in the skins of grapes, resveratrol is more concentrated in red wine and is often highlighted for its potential health benefits, including cardiovascular protection and anti-aging effects.</li>
        <li><strong>Quercetin</strong>: A flavonoid known for its antioxidant properties, quercetin contributes to the color and flavor of wine and may have anti-inflammatory effects, potentially reducing the risk of chronic diseases.</li>
        <li><strong>Catechins</strong>: These are flavan-3-ols found in both red and white wine, though more abundant in red wine. Catechins are potent antioxidants that may contribute to heart health by improving blood vessel function and reducing blood pressure.</li>
        <li><strong>Anthocyanins</strong>: Responsible for the red, purple, and blue colors in red wine, anthocyanins are powerful antioxidants that contribute to the taste profile and may offer protection against oxidative stress and inflammation.</li>
        <li><strong>Tannins</strong>: Tannins are a group of polyphenols that affect the astringency, structure, and aging potential of wine. Derived from grape skins, seeds, and stems, as well as from oak barrels used in aging, tannins contribute to the complexity and longevity of wine.</li>
        <li><strong>Procyanidins</strong>: A type of condensed tannin found in red wine, procyanidins are associated with cardiovascular benefits due to their ability to improve blood vessel function and reduce the risk of heart disease.</li>
        <li><strong>Gallic Acid</strong>: A simple phenolic compound, gallic acid contributes to the antioxidant capacity of wine. It is found in both red and white wines but is more prevalent in red wines, where it plays a role in preserving the wine's quality over time.</li>
    </ul>
    <p>These polyphenols are responsible for wine's antioxidant properties, color, and taste, making them essential for both the sensory experience and potential health benefits of wine.</p>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE

# Load the wine dataset
wine = load_wine()
X = wine.data
y = wine.target

# Define the two feature sets (corrected indices)
health_promoting_features = [9, 10, 11, 6, 5]  # indices of health-promoting features

# feature names for reference
print("Wine health-promoting features:")
for idx in health_promoting_features:
    print(f"- {wine.feature_names[idx]}")

# Function to perform t-SNE and plot results - this allows us to view the high-dimensional data in a 2D plot
def plot_tsne(X, y, features, title):
    # Select and scale the features
    X_selected = StandardScaler().fit_transform(X[:, features])
    
    # Perform t-SNE
    tsne = TSNE(n_components=2, random_state=42)
    X_tsne = tsne.fit_transform(X_selected)
    
    # Plot the results
    plt.figure(figsize=(6, 5))
    scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap='viridis')
    plt.title(title)
    plt.xlabel('t-SNE feature 1')
    plt.ylabel('t-SNE feature 2')
    plt.colorbar(scatter, label='Wine Class')
    plt.show()

# Plot t-SNE for health-promoting features
plot_tsne(X, y, health_promoting_features, 't-SNE of Wine Dataset: Health-Promoting Features')
```
